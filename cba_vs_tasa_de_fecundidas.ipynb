```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
niños = pd.read_excel(r"C:\Users\jaria\Desktop\tasa_de_natalidad.xlsx")
```

```python
niños
```

```python
niños
```

```python
niños = niños.drop(0).reset_index(drop=True)
```

```python
niños
```

```python
niños.info()
```

```python
niños["Diferencia %"] = niños["Tasa de natalidad"].pct_change() * 100
```

```python
niños
```

```python
niños = niños.fillna(0)
```

```python
niños["Diferencia %"] = niños["Diferencia %"].apply(lambda x: round(x, 3))
```

```python
niños
```

```python
fig, ax = plt.subplots()
ax.plot(niños["Año"], niños["Diferencia %"], "o-")
ax.grid()
ax.set_title("Cambios en % entre tasa de natalidad y salarios minimos")
ax.set_ylabel("Porcentaje de cambio (%)")
ax.set_xlabel("Años")
ax.set_xticks(range(2011, 2023, 1))
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(canasta["Año"],canasta["Diferencia %"], "o-", color="#2CA02C", label="Costo canasta básica")

ax.plot(niños["Año"], niños["Diferencia %"], "o-", color="#FF69B4", label="Tasa de natalidad")
ax.grid()
ax.set_title("Cambios en % entre costo de canasta básica y tasa de natalidad", fontsize=14)
ax.set_ylabel("Porcentaje de cambio (%)", fontsize=14)
ax.set_xlabel("Años", fontsize=14)

ax.legend(loc="upper left", fontsize=14)
for i in range(len(canasta["Año"])):
    ax.text(canasta["Año"][i], canasta["Diferencia %"][i] + 0.7, f"{round(canasta["Diferencia %"][i], 2)}%", size=13, va="bottom", ha="center")

for i in range(len(canasta["Año"])):
    ax.text(niños["Año"][i], niños["Diferencia %"][i] - 0.7, f"{round(niños["Diferencia %"][i], 2)}%", size=13, va="top", ha="center")

    
ax.margins(None, 0.2)
ax.set_xticks(range(2011, 2023, 1))
ax.tick_params(labelsize=14)
# ax.axhline(y=0, linestyle='-', color="k", alpha=0.4, linewidth=1, label='Línea horizontal en y=25')

fig.savefig("costo_natalidad.png", dpi=300, bbox_inches="tight")
```
